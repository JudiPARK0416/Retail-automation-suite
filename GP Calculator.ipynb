#1. Importing file

```python
import pandas as pd
from google.colab import drive

# 1. 구글 드라이브 마운트
drive.mount('/content/drive')

# 2. 경로 설정
file_path = "/content/drive/MyDrive/ColabData/24DEC.CSV"

# 3. CSV 불러오기
df = pd.read_csv(file_path, header=None)
df.dropna(how='all', inplace=True)
```

#2. Data Preprocessing

```python
# 2. 컬럼 구조 확인
print("🔍 원본 컬럼명:")
print(df.columns.tolist())

# 3. 컬럼명 재정의 (순서에 맞게 수동 지정)
# 데이터에 따라 다를 수 있으니 아래는 예시이며, 실제 컬럼 순서에 맞게 조정 가능
df.columns = [
    'Drop',            # 첫 번째 열은 필요 없어 보여요 (None)
    'Amount',
    'Quantity',
    'Unit Cost',
    'Total Amount',
    'Net',
    'GST',
    'GST2',
    'Net2',
    'GP%'

# 4. 불필요한 열 제거 (예: Drop, GST2, Net2 등)
df = df.drop(columns=['Drop', 'GST2', 'Net2'])

# 5. 인덱스 정리
df.reset_index(drop=True, inplace=True)

# 6. 결과 미리보기
print("✅ 정리된 컬럼:")
print(df.columns.tolist())
df.head()
```
